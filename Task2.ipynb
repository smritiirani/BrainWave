```python
from IPython.display import display, clear_output
import ipywidgets as widgets

users = {"admin": "admin123"}
inventory = {}

username_input = widgets.Text(description="Username:")
password_input = widgets.Password(description="Password:")
login_button = widgets.Button(description="Login", button_style='success')

product_name = widgets.Text(description="Name:")
product_qty = widgets.IntText(description="Quantity:")
product_price = widgets.FloatText(description="Price:")
add_button = widgets.Button(description="Add Product", button_style='primary')
edit_button = widgets.Button(description="Edit Product", button_style='warning')
delete_button = widgets.Button(description="Delete Product", button_style='danger')
view_inventory_button = widgets.Button(description="View Inventory")
low_stock_button = widgets.Button(description="Low Stock Alert")
output = widgets.Output()

def handle_login(b):
    clear_output()
    if username_input.value in users and users[username_input.value] == password_input.value:
        with output:
            print(f"Welcome, {username_input.value}!")
        show_inventory_ui()
    else:
        with output:
            print("Invalid username or password.")

def add_product(b):
    name = product_name.value
    qty = product_qty.value
    price = product_price.value
    if name and qty >= 0 and price >= 0:
        inventory[name] = {"quantity": qty, "price": price}
        with output:
            clear_output()
            print(f"Product '{name}' added.")
    else:
        with output:
            clear_output()
            print("Invalid input. Please check values.")

def edit_product(b):
    name = product_name.value
    if name in inventory:
        inventory[name]["quantity"] = product_qty.value
        inventory[name]["price"] = product_price.value
        with output:
            clear_output()
            print(f"Product '{name}' updated.")
    else:
        with output:
            clear_output()
            print(f"Product '{name}' not found.")

def delete_product(b):
    name = product_name.value
    if name in inventory:
        del inventory[name]
        with output:
            clear_output()
            print(f"Product '{name}' deleted.")
    else:
        with output:
            clear_output()
            print(f"Product '{name}' not found.")

def view_inventory(b):
    with output:
        clear_output()
        if inventory:
            for name, details in inventory.items():
                print(f"{name}: Quantity = {details['quantity']}, Price = ₹{details['price']:.2f}")
        else:
            print("Inventory is empty.")

def low_stock_alert(b):
    with output:
        clear_output()
        low_stock_items = {name: details for name, details in inventory.items() if details['quantity'] < 5}
        if low_stock_items:
            print("Low Stock Products (Qty < 5):")
            for name, details in low_stock_items.items():
                print(f"{name}: Quantity = {details['quantity']}")
        else:
            print("No low stock items.")

def show_inventory_ui():
    display(widgets.VBox([
        widgets.HBox([product_name, product_qty, product_price]),
        widgets.HBox([add_button, edit_button, delete_button]),
        widgets.HBox([view_inventory_button, low_stock_button]),
        output
    ]))

login_button.on_click(handle_login)
add_button.on_click(add_product)
edit_button.on_click(edit_product)
delete_button.on_click(delete_product)
view_inventory_button.on_click(view_inventory)
low_stock_button.on_click(low_stock_alert)

display(widgets.VBox([username_input, password_input, login_button, output]))
```
