```python
from IPython.display import display, clear_output
import ipywidgets as widgets

# Dummy ATM users
atm_users = {
    "1234": {"pin": "4321", "balance": 5000.0},
    "5678": {"pin": "8765", "balance": 10000.0}
}

current_user = None

# Login widgets
card_input = widgets.Text(description="Card No:")
pin_input = widgets.Password(description="PIN:")
login_button = widgets.Button(description="Login", button_style="success")
login_output = widgets.Output()

# Dashboard buttons
withdraw_button = widgets.Button(description="Withdraw")
deposit_button = widgets.Button(description="Deposit")
balance_button = widgets.Button(description="Check Balance")
logout_button = widgets.Button(description="Logout", button_style="danger")
dashboard_output = widgets.Output()

def show_login():
    clear_output()
    login_output.clear_output()
    display(card_input, pin_input, login_button, login_output)

def show_dashboard():
    clear_output()
    dashboard_output.clear_output()
    display(withdraw_button, deposit_button, balance_button, logout_button, dashboard_output)

def handle_login(btn):
    global current_user
    card = card_input.value.strip()
    pin = pin_input.value.strip()
    if card in atm_users and atm_users[card]["pin"] == pin:
        current_user = card
        show_dashboard()
    else:
        with login_output:
            clear_output()
            print("Invalid card number or PIN!")

def handle_logout(btn):
    global current_user
    current_user = None
    card_input.value = ""
    pin_input.value = ""
    show_login()

def handle_check_balance(btn):
    with dashboard_output:
        clear_output()
        print(f"Your balance is ₹{atm_users[current_user]['balance']}")

def handle_deposit(btn):
    amount_input = widgets.FloatText(description="Amount:")
    confirm_button = widgets.Button(description="Confirm Deposit", button_style="success")

    def confirm_deposit(b):
        amount = amount_input.value
        if amount > 0:
            atm_users[current_user]['balance'] += amount
            with dashboard_output:
                clear_output()
                print(f"₹{amount} deposited successfully!")
        else:
            with dashboard_output:
                clear_output()
                print("Enter a valid amount.")

    with dashboard_output:
        clear_output()
        display(amount_input, confirm_button)
        confirm_button.on_click(confirm_deposit)

def handle_withdraw(btn):
    amount_input = widgets.FloatText(description="Amount:")
    confirm_button = widgets.Button(description="Confirm Withdraw", button_style="warning")

    def confirm_withdraw(b):
        amount = amount_input.value
        if 0 < amount <= atm_users[current_user]['balance']:
            atm_users[current_user]['balance'] -= amount
            with dashboard_output:
                clear_output()
                print(f"₹{amount} withdrawn successfully!")
        else:
            with dashboard_output:
                clear_output()
                print("Insufficient balance or invalid amount.")

    with dashboard_output:
        clear_output()
        display(amount_input, confirm_button)
        confirm_button.on_click(confirm_withdraw)

# Button event bindings
login_button.on_click(handle_login)
logout_button.on_click(handle_logout)
balance_button.on_click(handle_check_balance)
deposit_button.on_click(handle_deposit)
withdraw_button.on_click(handle_withdraw)

# Start the app
show_login()
```
